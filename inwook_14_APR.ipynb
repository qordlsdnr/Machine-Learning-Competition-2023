```python
for i in range(1,10):

    plt.figure(figsize=(12,10))
    plt.suptitle(f'Training data # {i}')
    df=pd.read_csv(f'data/train/aligned_well_0{i}.csv')

    plt.subplot(1,4,1)
    plt.plot(df.GR, df.DEPT)
    plt.title("GR")
    plt.gca().invert_yaxis()
    plt.grid('on')

    plt.subplot(1,4,2)
    plt.plot(df.RHOB, df.DEPT)
    plt.title("RHOB")
    plt.gca().invert_yaxis()
    plt.grid('on')

    plt.subplot(1,4,3)
    plt.plot(df.NPHI, df.DEPT)
    plt.title("NPHI")
    plt.gca().invert_yaxis()
    plt.grid('on')
    
    plt.subplot(1,4,4)
    plt.plot(df.RD, df.DEPT)
    plt.title("RD")
    plt.gca().invert_yaxis()
    plt.xlim([0,100])

    plt.subplots_adjust(left=0.125, bottom=0.01, right=0.9, top=0.9, wspace=0.3, hspace=0.3)
    plt.savefig(f'Training data # {i}')

    plt.grid('on')
    plt.show()
```
